```python
def get_earth_B(r,fi):
    ## r in re;fi in degree
    r*=c.R_earth
    fi=fi/180*np.pi
    Me=8e22*u.A*u.m**2
    B=c.mu0*Me/(4*np.pi*r**3)
    fi=[-2*np.sin(fi),np.cos(fi)]
    B=B.to(u.gauss)

    return [B,fi]
```
