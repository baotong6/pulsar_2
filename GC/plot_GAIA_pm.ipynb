```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import pandas as pd
from astropy import units as u
from astropy import constants as c
from astropy.coordinates import SkyCoord,match_coordinates_sky
from astroquery.gaia import Gaia
from astropy.coordinates import Angle
```

```python
path='/Users/baotong/Desktop/period_terzan5/CSC/'
cat=pd.read_csv(path+'GAIA_NGC6397.csv')
```

```python
pmra=np.array(cat['pmra'])
pmdec=np.array(cat['pmdec'])
pmra = pmra[np.logical_not(np.isnan(pmra))]
pmdec = pmdec[np.logical_not(np.isnan(pmdec))]
```

```python
plt.title('NGC 6397')
plt.scatter(pmra-3.285,pmdec+17.621,s=5,color='red',marker='.')
plt.scatter(-1.351192684-3.285,1.632978057+17.621,s=100,color='green',marker='o')
plt.scatter(3.593983916-3.285,-16.87608252+17.621,s=100,color='green',marker='o')
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.xlabel('Proper motion in RA (mas/yr)')
plt.ylabel('Proper motion in DEC (mas/yr)')
plt.show()
```
